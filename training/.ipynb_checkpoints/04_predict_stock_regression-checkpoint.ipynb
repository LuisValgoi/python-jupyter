```python
import pandas as pd
import quandl
import math
import numpy as np
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# config quandl
quandl.ApiConfig.api_key = 'P74KecsqJ-S9PKGgsv18'

# get data & creating columns
df = quandl.get('WIKI/GOOGL')
df = df[['Adj. Open', 'Adj. High', 'Adj. Low', 'Adj. Close', 'Adj. Volume']]
df['VOLATILITY'] = (df['Adj. High'] - df['Adj. Low']) / df['Adj. Low'] * 100.0 
df['VARIATION'] = (df['Adj. Close'] - df['Adj. Open']) / df['Adj. Open'] * 100.0
df = df[['Adj. Close', 'VOLATILITY', 'VARIATION', 'Adj. Volume']]
df.dropna(inplace=True)

# choosing the column I want to predict
# selecting 10% of the dataframe to predict
# creating a label on top of the column w/ the negatives on top
forecast_col = 'Adj. Close'
forecast_out = int(math.ceil(0.01 * len(df)))
df['label'] = df[forecast_col].shift(-forecast_out)
df.dropna(inplace=True)

# define input and output
X = np.array(df.drop(['label'], 1))
y = np.array(df['label'])

# scale / standarized / equalize input set
X = preprocessing.scale(X)
y = np.array(df['label'])

# shuffle randomallicaly and use 20% of the data for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# predict models using the maximum of threads @ gpu
clf = LinearRegression(n_jobs=-1) 
clf.fit(X_train, y_train)
accuracy = clf.score(X_test, y_test)
print(f'The LinearRegression has an accuraccy of: {round(accuracy*100, 2)}% for {forecast_out} days in advance')
      
clf2 = svm.SVR(kernel='poly')
clf2.fit(X_train, y_train)
accuracy2 = clf2.score(X_test, y_test)
print(f'The Support Vector Regression has an accuraccy of: {round(accuracy2*100, 2)}% for {forecast_out} days in advance')
```
